# 네이버 환율 정보 크롤링하기

네이버 환율 정보를 직접 크롤링해봅시다!

- 크롤링 사이트 : <https://m.stock.naver.com/marketindex/index.nhn>

![img](https://aiffelstaticprd.blob.core.windows.net/media/original_images/exchange_rate.png)

- 국가와 환율 데이터를 크롤링합니다.
- 크롤링 한 데이터를 pandas의 Series 객체로 저장합니다.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://m.stock.naver.com/marketindex/index.nhn"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")
soup
```

```python
raw_list = soup.select("#marketindexLastList > li")

raw_list.pop(0)
currency = {}
for one in raw_list:
    try:
        country = one.select(".stock_item")[0].text
        price = one.select(".stock_price")[0].text
        currency[country] = price
    except:
        continue

print(currency)
```

```python
data = pd.Series(currency)
data
```
